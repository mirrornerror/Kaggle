## Kaggle: Digit Recognizer by Auto-Keras  
data: https://www.kaggle.com/c/digit-recognizer  
library: https://autokeras.com
  
reference: https://www.simonwenkel.com/2018/08/29/introduction-to-autokeras.html

```python
from autokeras.image_supervised import ImageClassifier
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
seed = 123
```

### Load MNIST(from Kaggle) data
* split data to train data and validation data  
* no normalization  

```python
train = pd.read_csv('../train.csv')
test = pd.read_csv('../test.csv')

label = train.label
test_index = test.index
train = train.drop(['label'], axis=1)

train = train.values.reshape(-1,28,28,1)
test = test.values.reshape(-1,28,28,1)
print(test.dtype)
label = label.astype('uint8')
train = train.astype('uint8')
test = test.astype('uint8')

from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(train, label, test_size = 0.2, random_state=seed)
print(x_train.shape, y_train.shape, x_val.shape, y_val.shape)
```

### visualize a sample image from the data:

```python
plt.imshow(x_val[0,:,:,0], cmap='gray')
```

### Train:

```python
clf = ImageClassifier(verbose=True, augment=True, searcher_args={'trainer_args':{'max_iter_num':5}})
clf.fit(x_train, y_train, time_limit=1 * 60 * 60) # 1 hours
```

### Search the Best model:

```python
clf.final_fit(x_train, y_train, x_val, y_val, retrain=False, trainer_args={'max_iter_num':10})
```

### Evaluate the result:

```python
y = clf.evaluate(x_val, y_val)
print(y)  
# 0.9873809523809524(result from one-hour-training)
# 0.9896428571428572(result from five-hour-training)
```

### Predict the submission test data:

```python
pred = clf.predict(test)
pred.shape
```

### Best model:  
output the number of layers 

```python
best_model = clf.load_searcher().load_best_model()
best_model.n_layers
```

### Save the best model:

```python
clf.load_searcher().load_best_model().produce_keras_model().save('autokeras_mnist_model.h5')
```

## Load the best model:
show the model summary

```python
from keras.models import load_model
model = load_model('autokeras_mnist_model.h5')
model.summary()
```

### Show the model diagram:

```python
from keras.utils import plot_model
plot_model(model, to_file='autokeras_mnist_model.png')
```
